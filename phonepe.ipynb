```python
#extract aggregate transaction data from json file

def agg_trans_data():
    path1 = "C:\\Users\\maria\\Downloads\\phonepe_data\\data\\aggregated\\transaction\\country\\india\\state\\"
    agg_trans_state_list=os.listdir(path1)

    tran_clm={'state':[],'year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

    for state in agg_trans_state_list:
        p_state=path1 + state + "/"
        agg_year=os.listdir(p_state)
        for year in agg_year:
            p_year=p_state+ year+"/"
            agg_Quater=os.listdir(p_year)
            for quater in agg_Quater:
                p_quater=p_year + quater
                data=open(p_quater,'r')
                d=json.load(data)
                for i in d['data']['transactionData']:
                    name=i['name']
                    count=i['paymentInstruments'][0]['count']
                    amount=i['paymentInstruments'][0]['amount']
                    tran_clm['Transaction_type'].append(name)
                    tran_clm['Transaction_count'].append(count)
                    tran_clm['Transaction_amount'].append(amount)
                    tran_clm['state'].append(state)
                    tran_clm['year'].append(year)
                    tran_clm['Quater'].append(int(quater.strip('.json')))

    df=pd.DataFrame(tran_clm)
    df['state']=df['state'].str.replace("-"," ").str.title()
    df['state']=df['state'].str.replace("Andaman & nicobar Islands","Andaman & Nicobar")
    df['state']=df['state'].str.replace('Dadra & Nagar Haveli & Daman & Diu','Dadra and Nagar Haveli and Daman and Diu')
        
    return(df)
```

```python
from sqlalchemy import create_engine, MetaData, Table, Column, String, Integer, Float, BigInteger
import pandas as pd

# Create a SQLAlchemy engine
engine = create_engine("mysql+mysqlconnector://root:root@localhost/phonepe")

# Create MetaData object
metadata = MetaData()

# Define tables using SQLAlchemy
agg_transaction = Table(
    'agg_transaction', metadata,
    Column('State', String(100)),
    Column('Year', Integer),
    Column('Quater', Integer),
    Column('Transacion_type', String(300)),
    Column('Transacion_count', BigInteger),
    Column('Transacion_amount', Float)
)

agg_user = Table(
    'agg_user', metadata,
    Column('State', String(100)),
    Column('Year', Integer),
    Column('Quater', Integer),
    Column('User_brand', String(50)),
    Column('User_count', BigInteger),
    Column('User_percentage', Float)
)

map_transaction = Table(
    'map_transaction', metadata,
    Column('State', String(100)),
    Column('District', String(100)),
    Column('Year', Integer),
    Column('Quater', Integer),
    Column('Transacion_count', BigInteger),
    Column('Transacion_amount', Float)
)

map_user = Table(
    'map_user', metadata,
    Column('State', String(100)),
    Column('District', String(100)),
    Column('Year', Integer),
    Column('Quater', Integer),
    Column('Registered_Users', BigInteger),
    Column('App_Opens', BigInteger)
)

top_transaction = Table(
    'top_transaction', metadata,
    Column('State', String(100)),
    Column('Pincode', Integer),
    Column('Year', Integer),
    Column('Quater', Integer),
    Column('Transacion_count', BigInteger),
    Column('Transacion_amount', Float)
)

top_user = Table(
    'top_user', metadata,
    Column('State', String(100)),
    Column('Pincode', Integer),
    Column('Year', Integer),
    Column('Quater', Integer),
    Column('Registered_Users', BigInteger)
)

# Create all tables in the database
metadata.create_all(engine)

# Define Data Insertion Function
def insert_data(table_name, data_function):
    data = data_function()
    data.to_sql(table_name, con=engine, if_exists='replace', index=False)

# Insert data using the defined function
insert_data('agg_transaction', agg_trans_data)
insert_data('agg_user', agg_user_data)
insert_data('map_transaction', map_trans_data)
insert_data('map_user', map_user_data)
insert_data('top_transaction', top_trans_data)
insert_data('top_user', top_user_data)

print("Data insertion completed successfully.")
```
